    INITIAL BASIC SEARCHING FOR INTERESTING DATA

```python
mediaEdad = st.mean(data.Age)
print("The average age of the players is: " + str(math.trunc(mediaEdad)) + " years")
```

```python
medianaEdad = st.median(data.Age)
print("The median age of the players is: " + str(math.trunc(medianaEdad)) + " years")
```

```python
bestPlayers = data.sort_values(['Overall'], ascending=0).head(5)
print("The best players in FIFA 19 are: ")
bestPlayers
```

```python
worstPlayers = data.sort_values(['Overall'], ascending=1).head(5)
print("The worst players in FIFA 19 are: ")
worstPlayers
```

```python
medianaOverall = st.median(data.Overall)
print("The median of the overall is: " + str(medianaOverall))

mediaOverall = st.mean(data.Overall)
print("The mean of the overall is: " + str(mediaOverall))
```

    Search for best potential with low price

```python
promises = data[(data["Potential"] > 90) & (data["Age"] < 28)]
promises['Release Clause'] = promises['Release Clause'].str.replace('€.', '', regex=True).replace('M', '', regex=True)
promises['Release Clause'] = promises['Release Clause'].astype('float')
promises
```

```python
coef = promises['Release Clause'] / promises['Potential']
bestPotentialValue = coef.sort_values(ascending=True).head(3)
```
