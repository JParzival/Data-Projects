    Search for best potential with low price

```python
promises = data[(data["Potential"] > 90) & (data["Age"] < 28)]
promises['Release Clause'] = promises['Release Clause'].str.replace('€.', '', regex=True).replace('M', '', regex=True)
promises['Release Clause'] = promises['Release Clause'].astype('float')
promises
```

```python
coef = promises['Release Clause'] / promises['Potential']
bestPotentialValue = coef.sort_values(ascending=True).head(3)
bestPotentialValue
```

CORRPLOT

```python
sns.set(style="white")

# Compute the correlation matrix
corr = data.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Some ideas can be extracted:

    1) All the GK abilities are proportionally inverse to the normal footballer abilities
    2) Each abilitie of the GK is slighty correlated to the rest
    3) The age slighty increases the Strength, the Composture and the Reactions, but is slighly correlated to a decrease in Acceleration, Sprint Speed and the Potential
    4) The strenght is inversely correlated to the balance
